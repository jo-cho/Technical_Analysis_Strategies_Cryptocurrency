## Stochastic Second Strategy

- Stochastic slow (14,3)
- long, short when K%  cross over D%(sma 20)

```python
window = 14

stochs = ta.momentum.StochasticOscillator(high,low,close,window)
```

```python
k = stochs.stoch_signal()
d = k.rolling(20).mean()
stocho = k-d
stocho.dropna(inplace=True)
```

```python
stocho
```

```python
a = np.sign(stocho).diff()*.5
crossover = a.loc[a!=0].dropna()
```
